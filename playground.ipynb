```python
from numpy import genfromtxt
sim_matrix = genfromtxt('sim_matrix.csv', delimiter=',')
feeds_titles = feeds.iloc[:,0].to_numpy()
```

```python
top10_dict = {} 
```

```python
for index, feed in enumerate(feeds_titles):
    sources_rank = []
    for i in range(sim_matrix.shape[0]):
        if i == index:
            continue
        sources_rank.append((feeds_titles[i], sim_matrix[index, i]))

    sources_rank.sort(key=lambda x: -x[1])
    
    top10_dict[feed] = [{'source':source[0], 'score':source[1]} for source in sources_rank[:10]]
```

```python
top10_dict
```

```python
with open('top10_similarity.json', 'w', encoding='utf-8') as f:
    json.dump(top10_dict, f, ensure_ascii=True, indent=4)
```
